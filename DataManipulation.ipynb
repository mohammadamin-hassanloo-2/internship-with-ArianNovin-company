```python
import pandas as pd #importing pandas library
df = pd.read_csv('data.csv') #setting our file to df as a DataFrame
```

```python
df.columns = df.iloc[0] #column names are not apropriate and first row is columns' name which is assigned! 
```

```python
column1 = df.iloc[:,0]
column2 = df.iloc[:,1]
column3 = df.iloc[:,2]
column4 = df.iloc[:,3]
column5 = df.iloc[:,4] #Each column is now a serie!
column1 = column1.str.strip()
column2 = column2.str.strip()
column3 = column3.str.strip()
column4 = column4.str.strip()
column5 = column5.str.strip() # stripped data is a clean data!
column1 = list(column1)
column2 = list(column2)
column3 = list(column3)
column4 = list(column4)
column5 = list(column5) # in order to join Series first I changed them! They are now lists!
df = pd.DataFrame(
    {
        'Year' :column1,
        'Month' : column2,
        'A' : column3,
        'B' : column4,
        'C' : column5
    }
) # Now it is a dataframe!
```
